```python
#--------
#INTRODUCTION TO DATA SCIENCE WORKFLOW#
#--------

#STEP1:import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#STEP2:load Dataset
df=pd.read_csv("Student_Performance_Dataset.csv")
#show first 5 rows
print(df.head())
```

```python
#STEP3:Understand Dataset
print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nDataset Information:")
print(df.info())
```

```python
plt.scatter(df["Study_Hours_Per_Day"], df["Final_Percentage"])
plt.xlabel("Study Hours")
plt.ylabel("Final Percentage")
plt.title("Do Study Hours Impact Performance?")
plt.show()
```
