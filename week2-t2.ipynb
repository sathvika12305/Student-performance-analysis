```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_excel("cleaned_titanic_dataset.xlsx")

# -----------------------------------
# 1. Survival Count Visualization
# -----------------------------------
survival = df["survived"].value_counts()

plt.figure(figsize=(6,5))
plt.bar(survival.index, survival.values)

plt.title("Passenger Survival Count")
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")
plt.xticks([0,1], ["Not Survived", "Survived"])

plt.show()

print("Insight:")
print("Most passengers did not survive the disaster.")

# -----------------------------------
# 2. Gender-wise Survival
# -----------------------------------
gender_survival = pd.crosstab(df["gender"], df["survived"])

gender_survival.plot(kind="bar", figsize=(7,5))

plt.title("Gender-wise Survival")
plt.xlabel("Gender")
plt.ylabel("Passenger Count")

plt.show()

print("Insight:")
print("Female passengers had a higher survival rate than males.")

# -----------------------------------
# 3. Passenger Class Distribution
# -----------------------------------
pclass = df["pclass"].value_counts()

plt.figure(figsize=(6,5))
plt.pie(
    pclass.values,
    labels=pclass.index,
    autopct="%1.1f%%"
)

plt.title("Passenger Class Distribution")

plt.show()

print("Insight:")
print("Most passengers belonged to third class.")

# -----------------------------------
# 4. Age Distribution
# -----------------------------------
plt.figure(figsize=(7,5))

plt.hist(df["age"], bins=15)

plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

print("Insight:")
print("Most passengers were young adults between 20 and 40 years.")

# -----------------------------------
# 5. Fare Distribution
# -----------------------------------
plt.figure(figsize=(7,5))

plt.hist(df["fare"], bins=15)

plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")

plt.show()

print("Insight:")
print("Most passengers paid lower ticket fares.")

# -----------------------------------
# 6. Survival by Passenger Class
# -----------------------------------
class_survival = pd.crosstab(df["pclass"], df["survived"])

class_survival.plot(kind="bar", figsize=(7,5))

plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()

print("Insight:")
print("First-class passengers had better survival chances.")

# -----------------------------------
# 7. Embarked Port Distribution
# -----------------------------------
embarked = df["embarked"].value_counts()

plt.figure(figsize=(6,5))

plt.bar(embarked.index, embarked.values)

plt.title("Embarked Port Distribution")
plt.xlabel("Port")
plt.ylabel("Passengers")

plt.show()

print("Insight:")
print("Most passengers boarded from Southampton (S).")

# -----------------------------------
# 8. Final Conclusion
# -----------------------------------
print("\nFINAL CONCLUSION")
print("1. Females had higher survival rates.")
print("2. First-class passengers survived more often.")
print("3. Young adults formed the majority of passengers.")
print("4. Most passengers paid lower fares and traveled in third class.")
print("5. Passenger class and gender strongly influenced survival.")
```
